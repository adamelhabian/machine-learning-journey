```python
import cv2
import matplotlib.pyplot as plt


# Task 2: Read Image
image = cv2.imread('newimage.jpg')
print("Image loaded successfully!")


# Task 3: Print Information
print("\nImage Information")
print("-" * 30)

print("Shape    :", image.shape)
print("Height   :", image.shape[0])
print("Width    :", image.shape[1])
print("Channels :", image.shape[2])
print("Data Type:", image.dtype)
print("Dimensions (ndim):", image.ndim)


# Task 4: Display RGB Image

rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

cv2.imshow('rgb_image',rgb_image)


# Task 5: Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('gray_image', gray_image)



# Task 6: HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
print("\nHSV Shape:", hsv_image.shape)


# Task 7: RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.imshow('rgb_image',rgb_image)



# Task 8: Resize to 300x200
small_image = cv2.resize(image, (300, 200))
print("New Shape:", small_image.shape)


# Task 9: Half Size (INTER_AREA)
half_image = cv2.resize(
    image,
    None,
    fx=0.5,
    fy=0.5,
    interpolation=cv2.INTER_AREA
)


# Task 10: Double Size (INTER_CUBIC)
large_image = cv2.resize(
    image,
    None,
    fx=2,
    fy=2,
    interpolation=cv2.INTER_CUBIC
)


# Task 11: Compare Interpolations
nearest = cv2.resize(
    image,
    None,
    fx=2,
    fy=2,
    interpolation=cv2.INTER_NEAREST
)

linear = cv2.resize(
    image,
    None,
    fx=2,
    fy=2,
    interpolation=cv2.INTER_LINEAR
)

cubic = cv2.resize(
    image,
    None,
    fx=2,
    fy=2,
    interpolation=cv2.INTER_CUBIC
)

fig, axes = plt.subplots(1, 3, figsize=(15,5))

axes[0].imshow(cv2.cvtColor(nearest, cv2.COLOR_BGR2RGB))
axes[0].set_title("INTER_NEAREST")
axes[0].axis("off")

axes[1].imshow(cv2.cvtColor(linear, cv2.COLOR_BGR2RGB))
axes[1].set_title("INTER_LINEAR")
axes[1].axis("off")

axes[2].imshow(cv2.cvtColor(cubic, cv2.COLOR_BGR2RGB))
axes[2].set_title("INTER_CUBIC")
axes[2].axis("off")

plt.show()


# Task 12: Save Images
cv2.imwrite("gray_image.jpg", gray_image)
cv2.imwrite("small_image.jpg", small_image)
cv2.imwrite("large_image.jpg", large_image)
print("\nImages Saved Successfully!")


# Task 13: Binary Image
thersh, binary_image = cv2.threshold(
    gray_image,
    127,
    255,
    cv2.THRESH_BINARY
)

cv2.imwrite("binary_image.jpg", binary_image)

print("Binary image saved successfully!")
```
