````python
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler,StandardScaler,Normalizer,Binarizer
from pandas import read_csv
from numpy import set_printoptions

path="pima-indians-diabetes.csv"
names=['preg','plas','pres','skin','test','mass','pedi','age','class']

# Data the inside file doesn't have columns names we assign column name to it using names attribute
dataframe=read_csv(path,names=names)
print(type(dataframe))
array=dataframe.values
print(type(array))
print(array.shape)

````

## Rows vs Columns in Preprocessing

### Normalizer
- Works on **rows (samples)**.
- Each **row is treated as one vector**.
- Scales each row so its **length (L2 norm) = 1**.

```text
Dataset

Height  Weight  Age
170      70      25   ← Vector (Row 1)
180      80      30   ← Vector (Row 2)
160      60      20   ← Vector (Row 3)
```

### Scalers (MinMaxScaler & StandardScaler)
- Work on **columns (features)**.
- Each feature is scaled independently.

```text
Height ↓
170
180
160

Weight ↓
70
80
60

Age ↓
25
30
20
```

### Easy Rule

- **Normalizer** → Normalize **Rows (Samples/Vectors)**.
- **MinMaxScaler** → Scale **Columns (Features)**.
- **StandardScaler** → Scale **Columns (Features)**.
- **Binarizer** → Converts each value independently to **0 or 1**.

## 1. Rescale Data (Min-Max Scaling)

- Scales all features to the same range (usually **0 to 1**).
- Improves the performance of many ML algorithms.
- Helps optimization algorithms like **Gradient Descent** converge faster.
- Important for algorithms sensitive to feature scale, such as:
  - Linear Regression
  - Neural Networks
  - K-Nearest Neighbors (KNN)

````python

#feature_range() -> normalize each feature so there values are between 0 and 1

# fit() -> scaler learns from dataset and get min and max of each feature 
# transform() -> applies the formula of the MINMAX SCALER 
data_scaler=MinMaxScaler(feature_range=(0,1))
data_rescaled=data_scaler.fit_transform(array)

set_printoptions(precision=2)

print("Rescaled Data :\n",data_rescaled[:5])
````

## 2. Normalization

- Scales each **row (sample)** to have a **length (norm) of 1**.
- Uses **L2 norm (Euclidean norm)** by default.
- Preserves the direction of the data.
- Different from Min-Max Scaling (0–1 scaling).
- Commonly used with KNN, clustering, and cosine similarity.

````python

data_normalizer=Normalizer(norm='l2').fit_transform(array)
set_printoptions(precision=3)
print(data_normalizer)
````

## 3. StandardScaler (Z-Score Scaling)

- Scales **columns (features)**.
- Makes each feature have:
  - **Mean = 0**
  - **Standard Deviation = 1**
- Improves model performance and training speed.
- Commonly used with:
  - Linear Regression
  - Logistic Regression
  - SVM
  - Neural Networks

**Equation**

$$
x'=\frac{x-\mu}{\sigma}
$$

Where:
- $x$ = original value
- $\mu$ = mean of the feature
- $\sigma$ = standard deviation of the feature
```

````python
data_scaler=StandardScaler().fit(array) # -> fit get mean and std
data_rescaled=data_scaler.transform(array)

set_printoptions(precision=2)

print("Rescaled Data :\n",data_rescaled[:5])
````

## Binarizer

- Converts numerical values into **binary values (0 or 1)**.
- Compares each value with a **threshold**.
- Values **greater than the threshold** become **1**.
- Values **less than or equal to the threshold** become **0**.
- Commonly used for binary classification and feature preprocessing.

**Equation**

$$
x' =
\begin{cases}
1, & x > \text{threshold} \\
0, & x \le \text{threshold}
\end{cases}
$$

Where:
- $x$ = original value
- **threshold** = cutoff value (default is `0.0`)
```

````python
binarizer=Binarizer(threshold=5)
data_binarized=binarizer.fit_transform(array)
print("Binarized Data : ",data_binarized )
````

## Data Imputation

- Replaces **missing values** with estimated values instead of deleting rows.
- Helps preserve the dataset and improve model performance.
- Common imputation methods applied on each **column**:
  - **Mean** (average) – for numerical data.
  - **Median** – for skewed numerical data.
  - **Mode** – for categorical data.

  - **Constant value** (e.g., 0 or "Unknown").


````python
import numpy as np
from sklearn.impute import SimpleImputer

data=[[1,2,np.nan,3],[1,2,3,np.nan],[1,2,np.nan,3],[1,2,np.nan,3]]

data=np.array(data)
print("Original Data : \n",data)
print("="*50)

# i can define missing_values as i want 
imp_mean=SimpleImputer(missing_values=np.nan , strategy='mean')
imp_mean.fit(data)
modified_data=imp_mean.transform(data)
print("modified_data using mean strategy : \n",modified_data)

print("="*50)

data=[[1,2,np.nan],[3,np.nan,1],[5,np.nan,0],[np.nan,4,6],[5,0,np.nan],[4,5,5]]

imp_const=SimpleImputer(missing_values=np.nan , strategy='constant',fill_value=100)
modified_data=imp_const.fit_transform(data)
print("Origianl Data : \n",np.array(data))
print("="*50)
print("modified_data using constant strategy :\n",modified_data)
````

# Example 4: Missing Value Imputation

> **Objective:** Load the Breast Cancer dataset and replace missing (`NaN`) values with the **mean** using `SimpleImputer`.

<details>
<summary><strong>Task</strong></summary>

- Load the Breast Cancer dataset.
- Create a matrix containing missing (`NaN`) values.
- Use **SimpleImputer(strategy="mean")**.
- Replace missing values with the column mean.
- Display the original and updated matrix.

</details>

````python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.impute  import SimpleImputer
import numpy as np

Breastdata=load_breast_cancer()

x=Breastdata.data
y=Breastdata.target

x[0,0] = np.nan
x[5,2] = np.nan

# Cleaning Data

imp=SimpleImputer(missing_values=np.nan,strategy='mean')
modified_data=imp.fit_transform(x)

print("Modified  Data : ",modified_data[:10])
print("y Data : ",y[:10])



#
````
