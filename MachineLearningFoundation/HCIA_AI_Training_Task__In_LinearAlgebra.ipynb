```python
import numpy as np
from numpy.linalg import norm
students = np.array([
    [20, 170, 72, 3.2, 6],
    [22, 165, 65, 3.8, 8],
    [21, 180, 78, 2.9, 5],
    [23, 175, 85, 3.5, 7],
    [24, 168, 70, 3.1, 6],
    [20, 172, 68, 3.6, 9],
    [25, 178, 90, 2.7, 4],
    [22, 169, 74, 3.9, 10]
])

print(students)
print("Shape:", students.shape)
print("Dimensions:", students.ndim)
print("Total Elements:", students.size)
print("Data Type:", students.dtype)

# Q3 dataset is matrix  because it has rows(students) and columns(features) 

print("First student :",students[0])
print("Last student :",students[-1])
print("Ages :",students[ : ,0])
print("GPA : ",students[ : ,3])
print("Heights and Weights :",students[ : , 1:3])


print("Average Age : ",np.mean(students[ : ,0]))
print("Average Height:", np.mean(students[:,1]))
print("Average Weight:", np.mean(students[:,2]))
print("Average GPA:", np.mean(students[:,3]))
print("Average Study Hours:", np.mean(students[:,4]))


print("Max Height : ",np.max(students[ : ,1]))
print("Minimum Weight:", np.min(students[:,2]))
print("Highest GPA:", np.max(students[:,3]))


print("Students whose GPA > 3.5",students[students[:,3] > 3.5])
print("Students who study >7 per week",students[students[:,4] >= 7])

age=students[:,0]
height=students[:,1]
weight=students[:,2]

print("Age - Height = ",age-height)
print("Height - Age =",height-age)
print("Height * Study Hours  = ",height*students[:,4])
print("Weight / Height = ",weight/height)

print("L1 Norm of weight :", norm(weight,1))
print("L2 Norm of weight :",norm(weight))
#Q21 Norm1 since it the sum of all absolute values while l2 square root of sum of squared values 

A=np.array([[4,7],2,6])
print("Determaint of A : ",np.linalg.det(A))
print("Inverse of A :",np.linalg.inv(A))
# yes it can be inversed sice it's determinat = 10 and not equal to 0

B = np.array([[1,2],
              [2,4]])

print("Determaint of B",np.linalg.det(B))

eigenvalues, eigenvectors = np.linalg.eig(A)

print(eigenvalues)
eigenvalues, eigenvectors = np.linalg.eig(A)

print(eigenvectors)

#Q27 Eigenvalues indicate how much data is scaled along a direction,
#  while eigenvectors define those directions. In Machine Learning
# , they are widely used in Principal Component Analysis (PCA) to identify the most important directions of variance in the data.
```
