#  Imaginary (Complex) Numbers

## What are they?
A complex number has two parts:
- **Real part** → x-axis
- **Imaginary part** → y-axis
It is written as:
\[
z = a + bi
\]
where:
- **a** = Real part
- **b** = Imaginary part
## Why are they important?
Complex numbers are used in:
- Signal Processing
- Fourier Transform
- Electrical Engineering
- Artificial Intelligence mathematics

````python
import numpy as np
import matplotlib.pyplot as plt

# Task:
# create the complex number, print it, plot it on a 2D graph, draw a vector from the origin (0, 0) to the complex number
# , calculate and display its magnitude and phase (in degrees), and add a title, grid, x-axis label, and y-axis label to the graph.

real=4
imaginary=3
z=4+3j

magnitude=np.sqrt(real**2 + imaginary**2)
mag=np.abs(z)
print(f"Manual Magnitude = {magnitude}")
print(f"Magnitude = {mag}")
print(f"Phase angle : {np.degrees(np.angle(z))}")
print(z)

plt.plot([0,real],[0,imaginary])
plt.grid(True,alpha=0.5)
plt.scatter(real,imaginary)
plt.xlabel("real")
plt.ylabel("imaginary")
plt.text(real, imaginary, " z")
plt.title("Complex Number")
````

# Probability Mass Function (PMF)

## What is it?
A **Probability Mass Function (PMF)** gives the probability of **each possible value** of a **discrete random variable**.
### Common ML Uses
- Class probabilities
- Naive Bayes
- Modeling discrete data
---
## Discrete Random Variable
A discrete random variable has **countable values**.
**Examples:**
- Dice outcome
- Number of students
- Number of emails
---
## Example: Fair Die
| Outcome (x) | P(X = x) |
|------------:|---------:|
|1|1/6|
|2|1/6|
|3|1/6|
|4|1/6|
|5|1/6|
|6|1/6|
The PMF answers:
> **"What is the probability that X equals a specific value?"**
---
## PMF Properties
- Every probability is between **0 and 1**.
- The sum of all probabilities is **1**.
---
## Visualization
A PMF is usually shown with a **bar chart**, because discrete values are separate.
---
## PMF vs Histogram

| PMF | Histogram |
|-----|-----------|
|Shows probabilities.|Shows frequencies (counts).|
|Represents a probability distribution.|Represents observed data.|
|Used for discrete variables.|Built from collected data.|

### Example

After rolling a die **600 times**:

- **Histogram:** Shows how many times each face appeared.
- **PMF:** Shows the theoretical probability of each face (**1/6**).


````python

# Task:
# A bag contains balls with the following numbers:
# [1, 1, 2, 2, 2, 3, 3, 4]
#
# 1. Store the numbers in a NumPy array.
# 2. Count how many times each number appears.
# 3. Calculate the probability (PMF) of each number.
# 4. Print the result as:
#
#    Number    Probability
#
# 5. Plot the PMF using a bar chart.
# 6. Add a title, x-axis label, y-axis label, and grid.
# 7. Verify that the sum of all probabilities is 1.

bag=[1, 1, 2, 2, 2, 3, 3, 4]
nums=[]
count=[]

for item in bag:
    if(np.isin(item,nums)==False):
        nums.append(item)

for i, num in enumerate(nums):
    for item in bag:
        if(num==item):
            if(len(count)<=i):
                count.append(1)
            else:
                count[i]=count[i]+1 
         
# print(nums)     
# print(count)   
p=[]
for i,item in enumerate(nums):
    print(f"Num {nums[i]} appear : {count[i]} Time")
    p.append(count[i]/sum(count))


print("    Number      Probability")  
for i in range(len(nums)):
    print(f"-    {nums[i]}           {p[i]}")

print(f"Total Sum Of Probabilty : {sum(p)}")
plt.bar(nums,p)
plt.xlabel("Nums")
plt.ylabel("Probability")
plt.title("PMF")
plt.grid(axis='y',alpha=0.3)
````

# Probability Density Function (PDF)

## What is it?

A **Probability Density Function (PDF)** describes the probability distribution of a **continuous random variable**.

### Examples
- Height
- Weight
- Temperature

### Key Points
- Used for **continuous data**.
- Probability is calculated over a **range (interval)**.
- The **total area under the curve is 1**.
- Represented by a **smooth curve**.

## PDF vs PMF

| PDF | PMF |
|-----|-----|
| Continuous data | Discrete data |
| Smooth curve | Bar chart |
| Probability over a range | Probability of an exact value |

> **Remember:** PMF counts, PDF measures.

````python
# Task:
# Create 200 x-values from -5 to 5 using NumPy.
# Calculate y = x^2.
# Plot the curve using plt.plot().
# Add a title, x-axis label, y-axis label, and grid.
# Observe how increasing the number of x-values makes the curve smoother.
#
# Bonus:
# Change the equation to y = x**3 or y = np.sin(x) and compare the curves.

x=np.linspace(-3,3,999)
y=x**2

plt.subplot(1,3,1)
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("PDF-> Y=x**2")
plt.grid(alpha=0.3)

y=x**3
plt.subplot(1,3,2)
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("PDF-> Y=x**3")
plt.grid(alpha=0.3)

y=np.sin(x)
plt.subplot(1,3,3)
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("PDF-> Y=sin(x)")
plt.grid(alpha=0.3)
plt.tight_layout()#adjust spacing between titles or labes of subplots

````

# Derivative & Integration

## Derivative
Measures **how fast a value changes** (the slope of a curve).

**ML Uses**
- Gradient Descent
- Model optimization
- Neural network training

> **Remember:** Derivative = Rate of change.

---

## Integration
Calculates the **total accumulated value** or the **area under a curve**.

**ML Uses**
- Probability Density Functions (PDFs)
- Expected value

> **Remember:** Integration = Area under the curve.

---

## Difference

| Derivative | Integration |
|------------|-------------|
| Rate of change | Total accumulation |
| Slope of a curve | Area under a curve |

**Easy way to remember:**
-  **Derivative** → *How fast?*
-  **Integration** → *How much?*
### `np.gradient()`

Calculates the **approximate derivative** (rate of change) of data.

```python
dy = np.gradient(y, x)
```

> **ML Use:** Finding gradients and analyzing data.
### `plt.fill_between()`

Fills the **area under a curve** or **between two curves**.

```python
plt.fill_between(x, y)
```

Or for a specific interval:

```python
plt.fill_between(x, y, where=(x >= a) & (x <= b))
```

> **ML Use:** Visualizing probability (area under a PDF).


````python
# Task:
# Consider the function y = x².
#
# 1. Create x-values from -5 to 5 using np.linspace().
# 2. Calculate y = x².
# 3. Plot the function.
# 4. Draw and shade the area under the curve between x = 0 and x = 3 (Integration).
# 5. Calculate the derivative y' = 2x.
# 6. Plot the derivative on the same figure with a different color.
# 7. Add a title, labels, legend, and grid.

x=np.linspace(-5,5,100)
y=x**2
y_dash=np.gradient(y,x)

plt.plot(x,y,label="y = x²")

plt.plot(x,y_dash,c="r",label="Derivative (dy/dx)")
plt.fill_between(x,y,where=(x>=0)&(x<=3),label='Area (0 ≤ x ≤ 3)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=x^2")

# Draw x-axis and y-axis through the origin
plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.xlim(-5, 5)
plt.ylim(-5, 30)
plt.grid(alpha=0.4)
plt.legend()

````

### Detecting Outliers

An **outlier** is a data point that is much different from the rest of the data.

**Common methods:**
- **IQR:** Values below `Q1 - 1.5 × IQR` or above `Q3 + 1.5 × IQR`.
- **Box Plot:** Points outside the whiskers are outliers.
- **Isolation Forest:** ML algorithm for detecting outliers in large datasets.

#  Box Plot (Box-and-Whisker Plot)

Summarizes numerical data and helps detect **outliers**.

### Shows
- Minimum
- Q1 (25%)
- Median (50%)
- Q3 (75%)
- Maximum
- Outliers

### IQR
```text
IQR = Q3 - Q1
```

Outliers:

```text
< Q1 - 1.5×IQR
> Q3 + 1.5×IQR
```

### Use For
- Finding the median
- Detecting outliers
- Comparing distributions
```

````python

scores=np.random.normal(70,10,200)



plt.figure(figsize=(6, 5))

plt.boxplot(
    scores,
    patch_artist=True,
    boxprops=dict(facecolor="skyblue"),
    medianprops=dict(color="red", linewidth=2)
)

plt.title("Distribution of Scores")
plt.ylabel("Score")
plt.grid(axis="y", alpha=0.3)

plt.show()
````

# Histogram, Density Histogram, and Bar Chart

## 1. Histogram

A **Histogram** is a graph that shows the **distribution of numerical (continuous) data** by grouping values into **bins (intervals)**.

### Characteristics
- X-axis: Numerical value ranges (bins)
- Y-axis: Frequency (count)
- Bars **touch** because the data is continuous.

### Uses
- Understand data distribution
- Find the center and spread
- Detect skewness
- Detect outliers
- Check if data is approximately normal

### Matplotlib

```python
plt.hist(data, bins=10)
```

---

## 2. Density Histogram

A **Density Histogram** is a normalized histogram that shows the **probability density** instead of the frequency.

Instead of counting observations, it estimates how the data is distributed.

### Characteristics
- X-axis: Numerical value ranges (bins)
- Y-axis: Density
- Bars **touch**
- Total **area under all bars = 1**
- Area of each bar represents the probability of that interval

### Uses
- Compare datasets with different sizes
- Estimate probability distributions
- Overlay probability density curves (e.g., Normal Distribution)

### Matplotlib

```python
plt.hist(data, bins=10, density=True)
```

---

## 3. Bar Chart

A **Bar Chart** compares the values of different **categories**.

Each bar represents one category.

### Characteristics
- X-axis: Categories
- Y-axis: Count or value
- Bars have **gaps** between them because categories are separate.

### Uses
- Compare categories
- Show counts or totals
- Display survey results
- Compare sales, populations, or frequencies

### Matplotlib

```python
plt.bar(categories, values)
```

---

# Comparison

| Feature | Histogram | Density Histogram | Bar Chart |
|---------|-----------|-------------------|-----------|
| Data Type | Numerical | Numerical | Categorical |
| Y-axis | Frequency | Density | Count or Value |
| Bars Touch? | ✅ Yes | ✅ Yes | ❌ No |
| Shows | Distribution | Probability Distribution | Category Comparison |
| Total Area = 1 | ❌ No | ✅ Yes | ❌ No |

---

# When to Use

### Histogram
- Explore numerical data

- Understand data distribution

### Density Histogram
- Compare datasets of different sizes

### Bar Chart

# Central Limit Theorem (CLT)

The **Central Limit Theorem (CLT)** states that:

> The distribution of **sample means** becomes approximately **normal** as the sample size increases, even if the original population is not normally distributed.

### Steps
1. Take many random samples.
2. Compute the mean of each sample.
3. Plot the sample means.
4. The distribution of the means becomes approximately normal.

### Conditions
- Random sampling
- Independent observations
- Sample size is usually **30 or more**

### Why it Matters
- Makes statistical inference possible.
- Used in confidence intervals and hypothesis testing.
- Widely applied in machine learning and data analysis.
## Standard Error (SE)

**Standard Error (SE)** measures how much **sample means** vary around the **population mean**.

### Formula

```text
SE = σ / √n
```

- **σ** = Population standard deviation
- **n** = Sample size

### Key Points

- Larger **sample size (n)** → Smaller **SE**
- Smaller **SE** → More accurate estimate of the population mean
- **Standard Deviation (SD):** Spread of individual data points
- **Standard Error (SE):** Spread of sample means

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Non-normal population
population = np.random.exponential(scale=2, size=100000)

sample_means = []

# Take 1000 samples
for _ in range(1000):
    sample = np.random.choice(population, size=100)
    sample_means.append(np.mean(sample))

mean=np.mean(sample_means)
se=np.std(sample_means)
SEmanual=np.std(sample)/np.sqrt(len(sample))
print(se)
print(SEmanual)
# Plot
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.hist(population, bins=40)
plt.title("Population")

plt.subplot(1,2,2)
plt.hist(sample_means, bins=30)
plt.axvline(mean,c="r",ls="--")
plt.axvline(mean-se,c='g',ls="--")
plt.axvline(mean+se,c='g',ls="--")
plt.title("Sample Means (CLT)")

plt.tight_layout()
````

# Normal Distribution

A **Normal Distribution** is a symmetric, bell-shaped probability distribution where most values are close to the mean.

### Equation

$$
f(x)=\frac{1}{\sigma\sqrt{2\pi}}
e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

### Parameters

- **μ (Mean):** Center of the distribution.
- **σ (Standard Deviation):** Controls the spread.

### Properties

- Bell-shaped and symmetric.
- Mean = Median = Mode.
- Total area under the curve = **1**.
- About **68%** of data lies within **±1σ**.
- About **95%** of data lies within **±2σ**.
- About **99.7%** of data lies within **±3σ**.

### Uses

- Modeling real-world data.
- Calculating probabilities.
- Statistical inference.
- Machine Learning.

````python

import numpy as np

np.random.seed(42)
data = np.random.normal(70, 10, 1000)
mean=np.mean(data)
std=np.std(data,ddof=1)

std1=data[(data>=(mean-std)) & (data<=(mean+std))]
std2=data[(data>=(mean-2*std)) & (data<=(mean+2*std))]
std3=data[(data>=(mean-3*std)) & (data<=(mean+3*std))]


print(f"Within 1σ: {(len(std1)/len(data))*100:.2f}%")
print(f"Within 2σ: {(len(std2)/len(data))*100:.2f}%")
print(f"Within 3σ: {(len(std3)/len(data))*100:.2f}%")

plt.figure(figsize=(6,4))
plt.hist(data,density=True)
plt.title("Normal Distruibution")
plt.xlabel("data")
plt.ylabel("Density")
plt.axvline(ls="-",c="black")
plt.axvline(mean,c="r",ls="--",label="mean")
plt.axvline(mean+std,c="y",ls="--",label="Mean + 1σ")
plt.axvline(mean-std,c="y",ls="--",label="Mean - 1σ")
plt.axvline(mean-2*std,c="g",ls="--",label="Mean - 2σ")
plt.axvline(mean+2*std,c="g",ls="--",label="Mean + 2σ")
plt.grid(alpha=0.3)
plt.legend()


plt.figure(figsize=(6,4))
plt.boxplot(data,showmeans=True)
plt.title("BoxPlot Of Data")

population = np.random.exponential(scale=2, size=100000)



def MakeSamples(size):
    sample_means = []

    for _ in range(1000):
        sample = np.random.choice(population, size=size)
        sample_means.append(np.mean(sample))

    return sample_means

sample_means=MakeSamples(30)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(population,density=True)
plt.axvline(np.mean(sample_means),label=f"Mean :{np.mean(population):.2f}",c="r",ls="--")
plt.legend()


sample30 = MakeSamples(30)

plt.hist(sample30)
plt.axvline(np.mean(sample30))
plt.subplot(1,2,2)
plt.hist(sample_means,density=True)
plt.axvline(np.mean(sample_means),label=f"Mean :{np.mean(sample30):.2f}",c="r",ls="--")
plt.legend()

sample5 = MakeSamples(5)
sample30 = MakeSamples(30)
sample100 = MakeSamples(100)
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.hist(sample5,density=True)
plt.axvline(np.mean(sample_means),label=f"Mean :{np.mean(MakeSamples(5)):.2f}",c="r",ls="--")
plt.legend()

plt.subplot(1,3,2)
plt.hist(sample30,density=True)
plt.axvline(np.mean(sample_means),label=f"Mean :{np.mean(MakeSamples(30)):.2f}",c="r",ls="--")
plt.legend()

plt.subplot(1,3,3)
plt.hist(sample100,density=True)
plt.axvline(np.mean(sample_means),label=f"Mean :{np.mean(MakeSamples(100)):.2f}",c="r",ls="--")
plt.legend()



````
