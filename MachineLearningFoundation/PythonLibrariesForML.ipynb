## 1.1`plt.plot()`

### What is it?
Draws a line by connecting data points.
It pairs each x-value with its corresponding y-value.

### Why use it?
Used to visualize trends or changes in data.

### Common ML Use
- Training loss
- Accuracy over epochs
- Predictions vs actual values

```python
# Draw a line in a diagram from position (0,0) to position (6,250):
# length of x-axis must equal to length of y-axis
# If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.
# we can use more param string shortcut 'o' to draw only the markers which mean 'rings'
#You can use the keyword argument marker to emphasize each point with a specified marker 
# format strings fmt shortcut string notation parameter to specify the marker  marker|line|color -> 'o:r' circed markers with dashed line all with red color
# You can use the keyword argument markersize or the shorter version, ms to set the size of the markers
#You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers
#You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers
#You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line
#You can use the keyword argument color or the shorter c to set the color of the line
#You can use the keyword argument linewidth or the shorter lw to change the width of the line
#You can draw multiple lines in same function plt.plot(x1, y1, x2, y2)
#With Pyplot, you can use the grid() function to add grid lines to the plot
#  You can use the axis parameter in the grid() function to specify which grid lines to display.
#Legal values are: 'x', 'y', and 'both'. Default value is 'both'.
#To change the transparency of the grid, use the alpha parameter in plt.grid()


ypoints = np.array([0, 250])
x=np.array([1, 2, 3, 4,5,6])
y=np.array([5, 7, 6, 8, 10,11])

#plt.plot(x,ms=10,mec='r',mfc='r',marker='o',ls='-',c='g',lw=10)
plt.plot(x,y,marker='o')
plt.grid(axis='both',c='g',ls='--',lw=0.5,alpha=0.5)
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales Trend")
plt.show()
```

## 1.2`plt.subplot()`

### What is it?
Divides a figure into multiple smaller plots.

### Why use it?
Display multiple graphs in one figure for comparison.

### Common ML Use
- Compare training vs validation loss.
- Compare different features.
- Compare different models.

```python
# You can add a title to the entire figure with the suptitle() function
#Every plt.plot() after subplot() is drawn inside the currently selected subplot.

#plot 1:
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])

plt.subplot(1,2,1)
##the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x1,y1)

#plot 2:
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])

plt.subplot(1,2,2)
plt.plot(x2,y2)




```

## 1.3`plt.scatter()`

### What is it?
Draws individual data points on a graph.

### Why use it?
Shows the relationship between two variables.

### Common ML Use
- Visualize datasets.
- Detect patterns.
- Find outliers.
- Check correlation between features.

```python
#The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis
#You can adjust the transparency of the dots with the alpha argument
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y,c='r',s=100)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y,alpha=0.4)

plt.show()
```

## 1.4`plt.bar()`

### What is it?
Creates a bar chart.

### Why use it?
Compares values between different categories.

### Common ML Use
- Compare class counts.
- Compare feature frequencies.
- Visualize categorical data.

```python
#If you want the bars to be displayed horizontally instead of vertically, use the barh() function
#If you want the bars to be displayed horizontally instead of vertically, use the barh() function
#The bar() takes the keyword argument width to set the width of the bars -> Thickness of the bars = heigth in barh
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(1,2,1)
plt.bar(x,y,color='red',width=0.9,)
plt.xlabel("Category")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.barh(x,y,height=0.9)

```

##1.6 `plt.pie()`

### What is it?
Creates a pie chart to show parts of a whole.

### Why use it?
Shows the percentage of each category.

### Common ML Use
- Class distribution.
- Dataset composition.
- Category proportions.

```python
#As you can see the pie chart draws one piece (called a wedge) for each value in the array
#By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:
#The size of each wedge is determined by comparing the value with all the other values, by using this formula:
#The value divided by the sum of all values: x/sum(x)
#The labels parameter must be an array with one label for each wedge
#The startangle parameter is defined with an angle in degrees, default angle is 0
#Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.
#The explode parameter, if specified, and not None, must be an array with one value for each wedge
#Add a shadow to the pie chart by setting the shadows parameter to True
#You can set the color of each wedge with the colors parameter.
# The colors parameter, if specified, must be an array with one value for each wedge
#To add a list of explanation for each wedge, use the legend() function
#To add a header to the legend, add the title parameter to the legend function

y=np.array([35,25,25,15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode)
plt.legend(title="Four Fruits")
```
