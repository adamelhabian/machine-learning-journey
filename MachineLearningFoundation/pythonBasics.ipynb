### `@staticmethod`

A **static method** is a method that belongs to a class but **does not need an object**.

- No `self` parameter.
- Uses only the input arguments.
- Called using the class name.

```python
class Math:

Math.add(2, 3)
```

**Use `@staticmethod` when the function doesn't use or modify object data and is only grouped inside the class for organization.**

A child class can replace a parent method.

class Dog(Animal):

class Dog(Animal):
