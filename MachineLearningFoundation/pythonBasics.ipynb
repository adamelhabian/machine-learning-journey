### **Generating Random Integers**

- **Purpose:** Generate random integer values within a specified range.
- **`random.randrange(start, stop)`**
  - Returns a random integer from **`start`** to **`stop - 1`**.
  - **The stop value is excluded.**
- **`random.randint(start, stop)`**
  - Returns a random integer from **`start`** to **`stop`**.
  - **The stop value is included.**
- **Use Case:** Useful for simulations, games, random sampling, testing, and machine learning experiments.

````python
import random

# End Excluded
print(random.randrange(1,10))
# End Included
print(random.randint(1,10))
````

### **List Filtering & Operations**

- `for` : Iterate through list elements.
- `if ... in` : Filter elements matching a condition.
- `append()` : Add an element to the end of a list.
- `sort()` : Sort the list in ascending order.
- `sort(reverse=True)` : Sort the list in descending order.
- `list()` : Create or copy a list.
- `copy()` : Create a shallow copy of a list.
- `+` : Concatenate two lists into a new list.

````python
fruits=["apple","banana","cherry","kiwi","mango"]
newlist=[]

for x in fruits:
    if "a" in x:
        newlist.append(x)
print(newlist)
newlist.sort()
print(newlist)
newlist.sort(reverse=True)
print(newlist)

nums=[133,2,66,3,77,0,10]
nums.sort()
print(nums)
nums.sort(reverse=True) # sort function applied only on the list of single data type 
print(nums)
newfruits=list(fruits)
print(newfruits)
copiedfruits=fruits.copy()
print(copiedfruits)
ls1=[1,2,3]
ls2=[4,5,6]
ls3=ls1+ls2
print(ls3)
````

### **Python Tuples**

#### **Properties**
- `()` : Create a tuple.
- Ordered : Elements keep their insertion order.
- Immutable : Items cannot be added, removed, or modified directly.
- Allows duplicate values.
- Supports multiple data types.
- Indexed : Access elements using positive or negative indices.
- Sliceable : Extract a range of elements using slicing.
- Iterable : Can be used in loops.
- Faster and more memory-efficient than lists.
- Useful for storing fixed (read-only) data.

#### **Important Functions & Methods**
- `len(tuple)` : Returns the number of elements.
- `type(tuple)` : Returns the object's data type.
- `max(tuple)` : Returns the largest value.
- `min(tuple)` : Returns the smallest value.
- `sum(tuple)` : Returns the sum of numeric elements.
- `tuple(iterable)` : Creates a tuple from an iterable.
- `count(value)` : Counts occurrences of a value.
- `index(value)` : Returns the index of the first matching value.
- `sorted(tuple)` : Returns a **sorted list** (does not modify the tuple).

#### **Ways to Edit a Tuple**

**1. Convert to a List (Recommended)**
```python
t = (1, 2, 3)
temp = list(t)
temp[1] = 20
t = tuple(temp)
print(t)
```

**2. Replace Using Slicing**
```python
t = (1, 2, 3, 4)

t = t[:1] + (20,) + t[2:]

print(t)
```
> **Note:** `(20,)` must include the comma because a single-item tuple requires a trailing comma.

#### **List vs Tuple**

| **List** | **Tuple** |
|----------|-----------|
| Uses `[]` | Uses `()` |
| Mutable | Immutable |
| Can add, remove, and update items | Cannot modify items directly |
| More methods available | Only `count()` and `index()` |
| Slightly slower | Slightly faster |
| Uses more memory | Uses less memory |
| Best for dynamic data | Best for fixed data |

### **Python Sets**

A **set** is an **unordered**, **mutable** collection of **unique** elements.

#### **Properties**
- Uses `{}` or `set()`.
- Unordered (no indexing or slicing).
- Does **not** allow duplicate values.
- Mutable (can add or remove items).
- Can store multiple data types.
- Fast for searching and membership testing (`in`).

#### **Important Functions & Methods**
- `len(set)` : Returns the number of elements.
- `type(set)` : Returns the object's data type.
- `add(value)` : Adds a new element.
- `update(iterable)` : Adds multiple elements from another iterable.
- `remove(value)` : Removes an element (raises an error if not found).
- `discard(value)` : Removes an element (no error if not found).
- `pop()` : Removes and returns a random element.
- `clear()` : Removes all elements.
- `copy()` : Creates a shallow copy.
- `union(set2)` / `|` : Returns all unique elements from both sets.
- `intersection(set2)` / `&` : Returns common elements.
- `difference(set2)` / `-` : Returns elements in the first set only.
- `symmetric_difference(set2)` / `^` : Returns elements that exist in only one set.

#### **List vs Tuple vs Set**

| **List** | **Tuple** | **Set** |
|----------|-----------|---------|
| Uses `[]` | Uses `()` | Uses `{}` |
| Ordered | Ordered | Unordered |
| Mutable | Immutable | Mutable |
| Allows duplicates | Allows duplicates | No duplicates |
| Supports indexing | Supports indexing | No indexing |
| Best for ordered data | Best for fixed data | Best for unique values |

````python
thisset={1,4,2,4,2}
s=set({4,3,5,3,5})
print(s)
s.add(66)
s.add("adam")
s.remove(5)
s.pop()

print(s)
#print(s[0]) give me error
print(thisset)
ss={111,1,2,3,5,6,7,8}
ss.pop()
ss.pop()
ss=ss.union({44,55,66,77,55})
print(ss)
````
